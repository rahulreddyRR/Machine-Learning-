# Matplotlib example workflow

```python
# Import matplotlib 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x=[1,2,3,4]
y=[10,20,30,40]

# 2. Setup the plot
fig, ax = plt.subplots(figsize=(10,10))

# 3.setplot
ax.plot(x,y)

# 3. Customize plot
ax.set(title="sample plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 4. save and show
fig.savefig("images/figs.png")
```

# Making figures with Numpy Array

```python
x = np.linspace(0,10,100)
x[:5]
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

```python
# Scattle Plot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# Bar plot
nut_butter_price ={"almend":7,
                  "Peanet":5,
                  "cashew":11}
fig,ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values());
ax.set(title="Butter Price",
       xlabel="list of Items",
       ylabel="Price ($)");
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()),list(nut_butter_price.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

### Two options of subplots

```python
# Option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_price.keys(),nut_butter_price.values());
ax4.hist(np.random.randn(1000));

ax1.set(title="Random");
```

```python
# subpolots in another methoed
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x**2);
ax[1,0].scatter(np.random.random(10),np.random.random(10));
```
